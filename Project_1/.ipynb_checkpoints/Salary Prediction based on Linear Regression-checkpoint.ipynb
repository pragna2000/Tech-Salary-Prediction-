# Linear Regression Model of Normalized Set

```python
real_df = pd.read_csv('../Data_CSVs/Real_Data.csv')
real_df
```

```python
y= real_df ['totalyearlycompensation']
X = real_df.drop(columns= ['totalyearlycompensation'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 42)
# X_train1 = train_df
# X_test = test_df

# y_train1 = train_df['totalyearlycompensation']
# y_test = test_df
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
from sklearn.linear_model import LinearRegression

# Create a linear regression model
model1 = LinearRegression()

# Fit the model to the training data
model1.fit(X_train, y_train)

# Print the coefficients and intercept of the model
print("Coefficients:", model1.coef_)
print("Intercept:", model1.intercept_)
```

```python
score = model1.score(X_test, y_test)
print("R-squared:", score)
```

```python
y_pred=model1.predict(X_test)
y_pred=y_pred.reshape(-1,1)
y_pred.shape
```

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y_test, y_pred))
print(metrics.mean_squared_error(y_test, y_pred))
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
plt.hist(y_test,bins = 100, alpha = 0.5)
plt.hist(y_pred, bins = 100, alpha = 0.5)
plt.show
```
