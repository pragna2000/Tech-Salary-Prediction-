# Data Cleaning

```python
# Removing unwanted columns from the data frame
data = df.drop(['job description','company_size', 'company_sector', 'company_industry', 'company_founded', 'company_revenue', 'Unnamed: 11', 'rating', 'description_len', 'company_age' ], axis =1)
data
```

```python
data.shape
```

```python
#renaming column titles
data= data.rename({'salary estimate':'salary'}, axis =1)
data
```

```python
# binary encoding of seniority 
sen =data['seniority'].replace({'mid':0,'junior':1, 'Senior':2})
data['seniority'] = sen
data
```

```python
data.dtypes
```

```python
sns.distplot(data['salary'])
```

```python
x =data[['seniority']]
x
```

```python
y = data[['salary']]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size =0.4)
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(x_train,y_train)
```

```python
y_pred = lr_model.predict(x_test)
```

```python
y_test['salary']
```

```python
result = {"Actual_salary": y_test['salary'], "prediction": y_pred}
```
